```python
sb.histplot(fuel_econ['co2'], bins=10, kde=False)
plt.title('CO2 Emission Distribution')
plt.xlabel('CO2 Emissions in grams per mile')
plt.ylabel('Frequency')
plt.show()
```

```python
sb.histplot(fuel_econ['drive'], bins='auto')
plt.title('Vehicle Drive type distribution')
plt.xlabel('Wheel drive type')
plt.ylabel('Frequency')
plt.show()
```

```python
sb.histplot(fuel_econ['fuelType'], bins='auto')
plt.title('Fuel type distribution')
plt.xlabel('Fuel type')
plt.ylabel('Frequency')
plt.show()
```

```python
data = pd.read_csv(path)
hm_data = data.drop(['make', 'model', 'VClass', 'drive', 'trans','fuelType'], axis=1)
correlation = hm_data.corr()
plt.figure(figsize=(12, 6))
sb.heatmap(correlation, vmin=-1, vmax=1, center=0, cmap=sb.diverging_palette(20, 220, n=500), square=True)
plt.xticks(rotation=45)
plt.show()
```
