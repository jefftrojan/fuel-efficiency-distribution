````python
#TO DO: Create a function that does a derivative for any function

from sympy import symbols, diff, sympify

def dy_dx(func_str, x_value):
    """
    Calculate the symbolic derivative of a function at a given point using SymPy.

    Parameters:
    - func_str: The function as a string.
    - x_value: The point at which to calculate the derivative.

    Returns:
    - The symbolic derivative of the function at the given point.
    """
    x = symbols('x')

    try:
        # Convert the input string to a SymPy expression
        func = sympify(func_str)

        # Calculate the derivative symbolically
        derivative_result = diff(func, x)

        # Substitute the x_value into the derivative result
        result_at_point = derivative_result.subs(x, x_value)

        return result_at_point

    except:
        return "Invalid input"
````

# Task 4: Create separate array of Minimas and maximas

````python
import sympy as sp

def find_minimas_maximas(func_str):
    x = sp.symbols('x')
    func = sp.sympify(func_str)

    # Calculate the derivative symbolically
    derivative_result = sp.diff(func, x)

    # Find critical points (where derivative is zero or undefined)
    critical_points = sp.solve(derivative_result, x)

    # Evaluate the second derivative to determine concavity
    second_derivative = sp.diff(derivative_result, x)

    arr_minimas = []
    arr_maximas = []

    for point in critical_points:
        # Use the second derivative test to determine if it's a minima or maxima
        if second_derivative.subs(x, point) > 0:
            arr_minimas.append(point)
        elif second_derivative.subs(x, point) < 0:
            arr_maximas.append(point)

    return arr_minimas, arr_maximas

def find_global_minima_maxima(func_str):
    x = sp.symbols('x')
    func = sp.sympify(func_str)

    # Calculate the derivative symbolically
    derivative_result = sp.diff(func, x)

    # Find critical points (where derivative is zero or undefined)
    critical_points = sp.solve(derivative_result, x)

    # Evaluate the function at critical points to find values
    function_values = [func.subs(x, point) for point in critical_points]

    # Find global minima and maxima
    global_minima = min(zip(critical_points, function_values), key=lambda x: x[1])
    global_maxima = max(zip(critical_points, function_values), key=lambda x: x[1])

    return global_minima, global_maxima


````

````python
# Example usage with the quadratic equation
input_function = "2*x**2 - 4*x + 3"
arr_minimas, arr_maximas = find_minimas_maximas(input_function)
global_minima, global_maxima = find_global_minima_maxima(input_function)

print(f'The array of minimas: {arr_minimas}')
print(f'The array of maximas: {arr_maximas}')
print(f'The global minima is at {global_minima}')
print(f'The global maxima is at {global_maxima}')
````
